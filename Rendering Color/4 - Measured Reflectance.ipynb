# Read measured reflectance of color chart

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Read data and separate each run

```python
data = pd.read_csv('act3_data.csv')
```

```python
run1 = data[6:644]
run2 = data[652:1290]
run3 = data[1298:1936]
run4 = data[1944:2582]
run5 = data[2590:3228]
run6 = data[3286:3874]
run7 = data[3882:4520]
run8 = data[4528:5166]
run9 = data[5174:5812]
run10 = data[5820:6458]
run11 = data[6466:7104]
run12 = data[7112:7750]
run13 = data[7758:8396]
run14 = data[8404:9042]
run15 = data[9050:9688]
```

```python
run16 = data[9696:10334]
run17 = data[10342:10980]
run18 = data[10988:11626]
run19 = data[11634:12272]
run20 = data[12280:12918]
run21 = data[12926:13564]
run22 = data[13572:14210]
```

## Plot reflectance of red-orange colors

```python
plt.figure(figsize=(8,6))

plt.plot(np.float_(run1['Vernier Format 2']),np.float_(run1['Unnamed: 1']),color='red',label='red')
plt.plot(np.float_(run7['Vernier Format 2']),np.float_(run7['Unnamed: 1']),color='palevioletred',label='dark salmon')
plt.plot(np.float_(run8['Vernier Format 2']),np.float_(run8['Unnamed: 1']),color='darkorange',label='orange')

plt.xlabel('wavelength (nm)',fontsize=14)
plt.ylabel('intensity (rel)',fontsize=14)
plt.legend(fontsize=12)
# plt.savefig('red.pdf',dpi=600,bbox_inches='tight')
plt.show()
```

## Plot reflectance of brown colors

```python
plt.figure(figsize=(8,6))

plt.plot(np.float_(run11['Vernier Format 2']),np.float_(run11['Unnamed: 1']),color='burlywood',label='light brown')
plt.plot(np.float_(run12['Vernier Format 2']),np.float_(run12['Unnamed: 1']),color='darkgoldenrod',label='nude')

plt.xlabel('wavelength (nm)',fontsize=14)
plt.ylabel('intensity (rel)',fontsize=14)
plt.legend(fontsize=12)
# plt.savefig('brown.pdf',dpi=600,bbox_inches='tight')
plt.show()
```
